```python
import os
```

```python
coords = []
with open('localMLESpace.log', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if not line.startswith('constrained'):
            spLine = line.split(',')
            lat, lon = spLine[0], spLine[1]
            coords.append([float(lat), float(lon)])
```

```python
with open('badcoords', 'w+') as f:
    for coord in coords:
        f.write(str(coord) + '\n')
```
